```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
X = np.random.rand(1000,1)
Y=3*X +4*np.random.randn(1000,1)
```

```python
Y
```

```python
plt.plot(X,Y,'b.')
```

```python
theta=np.random.rand(2,1)
```

```python
theta.shape
```

```python
X_mod=np.c_[np.ones((1000,1)),X]
```

```python
X_mod
```

```python
def hypothesis(X_mod,theta) :
    return X_mod.dot(theta)

```

```python
def cost_fn(X_mod,theta,Y):
    cost=(1/(2*1000)) * np.sum(np.square(hypothesis(X_mod,theta)-Y))
    return cost
```

```python
learning_rate=0.01
iterations=1000
```

```python
cost_history=[]
```

```python
def gradient_desc(X_mod,theta,Y,learning_rate,iterations):
    for iteration in range(iterations):
        error=hypothesis(X_mod,theta)-Y
        correction_term = (1/1000) * learning_rate * X_mod.T.dot(error)
        theta=theta-correction_term
        cost_history.append(cost_fn(X_mod,theta,Y))
        
```

```python
gradient_desc(X_mod,theta,Y,learning_rate,iterations)
```

```python
plt.plot(X,Y,'r.')
plt.plot(X,hypothesis(X_mod,theta))
```

```python
plt.plot(range(iterations),cost_history)
```

```python
predictions = X_mod.dot(theta)
```

```python
def MAE (Y, predictions) :
    return np.mean(np.abs(Y-predictions))
```

```python
MAE(Y,predictions)
```

```python
def MSE(Y,predictions):
    return np.mean(np.square(Y-predictions))
```

```python
MSE(Y,predictions)
```

```python
def r_squared(Y,predictions):
    num = np.sum(np.square(Y-predictions))
    den=np.sum(np.square(Y-np.mean(Y)))
    return (1 -(num/den)) * 100 
```

```python
r_squared(Y,predictions)
```
