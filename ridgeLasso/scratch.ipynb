```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
X=np.random.rand(10000,1)
Y=3*X+4*np.random.randn(10000,1)
```

```python
plt.scatter(X,Y)
```

```python
w=np.random.rand(1,1)
```

```python
b=np.random.rand()
```

```python
w,b
```

```python
m=10000
lamda=1
learning_rate=0.01
epochs=100000
```

```python
def cost_fn(X,w,b,Y):
    cost= (1/(2*m) * np.sum(np.square(X.dot(w) + b - Y))) + lamda/2 * np.sum(np.square(w)) 
    return cost
```

```python
cost_fn(X,w,b,Y)
```

```python
cost_history=[]
```

```python
def gradient(X,w,b,Y):
    for i in range(epochs):

        hype=X.dot(w) + b
        dw =  (1/m * X.T.dot(hype - Y )) + lamda * w
        db = 1/m * np.sum(hype-Y)

        w= w - learning_rate*dw
        db=b - learning_rate*dw
        cost=cost_fn(X,w,b,Y)
        cost_history.append(cost)


```

```python
gradient(X,w,b,Y)
```

```python
cost_history[0]
```

```python
plt.plot(range(epochs),cost_history)
```
