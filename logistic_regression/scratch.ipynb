```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
m=1000
```

```python
X=np.random.normal(loc=0,scale=1,size=(1000,1))
```

```python
Y=np.random.binomial(n=1,p=0.5,size=(1000,1))
```

```python
plt.plot(X,Y,'r.')
```

```python
w=np.random.randn(1,1)
```

```python
b=np.random.randn()
```

```python
def hype(X,w,b):
    h=X.dot(w)+b
    z=1 / (1+np.exp(-h))
    return z
```

```python
np.squeeze((hype(X,w,b)))
```

```python
def cost_fn(X,w,b,Y):
    cost= -(Y.T.dot(np.log(hype(X,w,b))) + (1-Y).T.dot(np.log(1-hype(X,w,b)))) * 1/m
    return cost
```

```python
learning_rate=0.1
epochs=10000
```

```python
cost_history=[]
```

```python
def gradient(X,w,b,Y,learning_rate,epochs):

    for i in range(epochs):
        dz=hype(X,w,b) - Y
        dw = 1/m * X.T.dot(dz)
        db = 1/m * np.sum(dz)
        w = w - learning_rate * dw
        b = b - learning_rate * db
        cost=cost_fn(X,w,b,Y)
        cost_history.append(cost)



```

```python
gradient(X,w,b,Y,learning_rate,epochs)
```

```python
np.squeeze(cost_history).shape
```

```python
plt.plot(range(epochs),np.squeeze(cost_history))
```
