```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
np.random.seed(0)
```

```python
X=np.random.rand(100,1)
# X.size
```

```python
Y=4+3*X+np.random.randn(100,1)
# Y
```

```python
#plotting the data
plt.plot(X,Y,'b.')
plt.xlabel('feature')
plt.ylabel('target')
```

```python
plt.scatter(X,Y)
plt.xlabel('feature')
plt.ylabel('target')
```

```python
n=100
learning_rate=0.01
iterations=1000
```

```python
theta=np.random.randn(2,1)
theta.shape
```

```python
X_b=np.c_[np.ones((n,1)),X]
```

```python
def compute_cost(X_b,Y,theta):
    m=len(Y)
    predictions=X_b.dot(theta)
    cost=(1/2*m)*np.sum(np.square(predictions-Y))
    return cost
```

```python
cost_history=[]
```

```python
X_b.shape
```

```python
for iteration in range(iterations):
    prediction=X_b.dot(theta)
    error=prediction-Y
    gradient=1/n*X_b.T.dot(error)
    theta=theta-learning_rate*gradient
    cost=compute_cost(X_b,Y,theta)
    cost_history.append(cost)

    
```

```python
plt.figure(figsize=(10, 6))
plt.plot(X, Y, "b.")
plt.plot(X, X_b.dot(theta), "r-", linewidth=2, label="Regression Line")
plt.xlabel("X")
plt.ylabel("Y")
plt.title("Linear Regression from Scratch")
plt.legend()
plt.show()
```

```python
cost_history.pop(0)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(range(iterations), cost_history, "b-")
plt.xlabel("Number of iterations")
plt.ylabel("Cost")
plt.title("Cost Function during Gradient Descent")
plt.show()
```
